```python
import pandas as pd
df=pd.read_csv('datapoints.csv')
```

```python
df=df.drop_duplicates()
```

```python
df.class_fruit.value_counts()
```

```python
y=df.class_fruit
df=df.drop('class_fruit',axis=1)
```

```python
df
```

```python
import sklearn
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(df,y,test_size=0.2,stratify=y)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
X_train=X_train/255
X_test=X_test/255
```

```python
X_train
```

```python
tree1_decision=DecisionTreeClassifier(max_depth=7,random_state=42)
```

```python
tree1_decision.fit(X_train,y_train)
```

```python
output=tree1_decision.predict(X_test)
```

```python
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test, output))
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.svm import SVC
from sklearn.ensemble import AdaBoostClassifier
```

```python
rnd_clf=RandomForestClassifier(n_estimators=100,max_leaf_nodes=5,n_jobs=-1,bootstrap=True,oob_score=True,max_samples=200)
```

```python
rnd_clf.fit(X_train,y_train)
```

```python
out=rnd_clf.predict(X_test)
print("Accuracy:",metrics.accuracy_score(y_test, out))
```

```python
ada_clf=AdaBoostClassifier(DecisionTreeClassifier(max_depth=3),n_estimators=200,learning_rate=0.01)
```

```python
ada_clf.fit(X_train,y_train)
res=ada_clf.predict(X_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, res))
```

```python
rnd_clf1=RandomForestClassifier(n_estimators=100,max_leaf_nodes=5,n_jobs=-1,bootstrap=True,oob_score=True,max_samples=200)
```

```python
knn=KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train,y_train)
pre=knn.predict(X_test)
print("Accuracy:",metrics.accuracy_score(y_test, pre))
```

```python
voting_clf=VotingClassifier(
    estimators=[('tree',tree1_decision),('rfc',rnd_clf),('ada_clf',ada_clf),('knn',knn),],
    voting='hard')
```

```python
voting_clf.fit(X_train,y_train)
res=voting_clf.predict(X_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, res))
```
